```python
import pandas as pd
df=pd.read_csv("spam.csv")
df
```

```python
df.groupby('Category').describe()
```

```python
#Convert Category into numbers
#Machine learning works better with numbers, not text labels.
#Spam = 1, Ham = 0 → this is called target encoding.

df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message,df.spam)
```

```python
#Convert text to numbers using CountVectorizer
#Machines cannot understand text. CountVectorizer converts words into numbers using a bag-of-words approach.

from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
#Predict new emails

emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
#Sklearn Pipeline

#Why use Pipeline?
#Combines vectorization + model in one step.
#Simplifies code → no need to manually transform text every time.
#Reduces errors → ensures same preprocessing is applied during training & testing.

from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
