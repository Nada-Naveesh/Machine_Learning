```python
import pandas as pd
df=pd.read_csv("titanic (1).csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
dummies = pd.get_dummies(inputs['Sex'])

dummies.head(3)

```

```python
#here we are concatinating two columns

inputs=pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
#Handle Missing Values

#Missing age values are replaced with average age.

#Naive Bayes cannot work with NaN values.


inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
#Gaussian Naive Bayes is used because:Data like Age and Fare are continuous fit() trains the model.

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test[0:10])
```

```python
#Calculate the score using cross validation

from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train, y_train, cv=5)
```
