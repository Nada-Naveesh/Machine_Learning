```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
digits
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5):
  plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(x_test)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.predict(x_test)
```

```python
plt.matshow(digits.images[56])
```

```python
reg.predict([digits.data[56]])
```

```python
#Cobnfusion Matrix
#A confusion matrix is a table used to evaluate the performance of a classification model.
#It provides a detailed breakdown of the model's predictions compared to the actual outcomes,
# helping to identify where the model is making errors.
#The matrix is particularly useful for binary and multi-class classification problems.

y_predicted = reg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (8,6))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
